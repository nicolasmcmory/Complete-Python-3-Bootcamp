# Object Oriented Programming Challenge

For this challenge, create a bank account class that has two attributes:

* owner
* balance

and two methods:

* deposit
* withdraw

As an added requirement, withdrawals may not exceed the available balance.

Instantiate your class, make several deposits and withdrawals, and test to make sure the account can't be overdrawn.

```python
from abc import ABC, abstractmethod


class Account(ABC):

    def __init__(
        self,
        owner_name: str = "undeclared",
        opening_balance: float = 0,
        account_limit: int = 500,
    ):
        self.owner = owner_name
        self.balance = opening_balance
        self.account_limit = account_limit

    # Dunder methods
    # To return a custom string of the Account instance
    def __str__(self):
        return f"Owner: {self.owner}\nBalance: {self.balance}"

    def deposit(self, amount: float) -> float:
        self.balance += amount
        print(f"Deposit acccepted.\n{self}")
        return self.balance

    def withdraw(self, amount: float) -> float:

        if self.balance - amount < 0:
            return print(f"Insuficient funds.\n{self}")

        else:
            self.balance -= amount
            print(f"Withdrawal acccepted.\n{self}")
            
        return self.balance
```

```python
# 1. Instantiate the class
acct1 = Account('Jose',100)
```

```python
# 2. Print the object
print(acct1)
```

```python
# 5. Make a series of deposits and withdrawals
acct1.deposit(50)
```

```python
acct1.withdraw(75)
```
