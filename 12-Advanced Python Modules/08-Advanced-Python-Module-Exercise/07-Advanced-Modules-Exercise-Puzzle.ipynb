```python
import zipfile
```

```python
# Unzipping intructions
zip_obj = zipfile.ZipFile("./unzip_me_for_instructions.zip", "r")
zip_obj.extractall("extracted_content")
```

```python
# Getting phone number
import os
import re
import pdb
import timeit

pattern = re.compile(r"\d{3}-\d{3}-\d{4}")
d = os.getcwd() + "/extracted_content"


def main_precompiled():
    for dir_path, sub_folders, files in os.walk(d):
        for f in files:
            filepath = os.path.join(dir_path, f)
            with open(filepath, "r") as f_open:
                text = f_open.read()
                result = re.search(pattern, text)
                if result:
                    return result.group()

def main_not_precompiled():
    for dir_path, sub_folders, files in os.walk(d):
        for f in files:
            filepath = os.path.join(dir_path, f)
            with open(filepath, "r") as f_open:
                text = f_open.read()
                result = re.search(r"\d{3}-\d{3}-\d{4}", text)
                if result:
                    return result.group()
```

```python
timeit.timeit("main_precompiled()",globals=globals(),number=100)
```

```python
timeit.timeit("main_not_precompiled()",globals=globals(),number=100)
```
